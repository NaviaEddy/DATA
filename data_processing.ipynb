# ***PROCESAMIENTO DE DATOS***

```python
#Renombramiento de columnas y verificacion de datos
Nuevos_nombres = {'PassengerId':'ID'}
data = data.rename(columns=Nuevos_nombres)
data.info()
```

```python
#Movimiento de columna al final del dataset y verificacion de datos
indice_columna = data.columns.get_loc('Survived')
columnas = list(data.columns)
columnas.pop(indice_columna)
columnas.append('Survived')
data = data[columnas]
data.info()
```

```python
#Cambio de los datos de tipo object a int64 y verificacion de datos
data['Sex'].replace(['female','male'],[0,1],inplace=True)
data['Embarked'].replace(['Q','S','C'],[0,1,2],inplace=True)
data.info()
print(data)
```

```python
#Sacamos los valores minimos y maximos de cada columna que necesite rellenar los datos
print("EDAD")
print(data['Age'].min())
print(data['Age'].max())
print("TARIFAS")
print(data['Fare'].min())
print(data['Fare'].max())
```

```python
#Reemplazamos todos los valores NaN de la columna 'Age', 'Fare'
for i in range(len(data)):
    if pd.isna(data.at[i, 'Age']):
        data.at[i, 'Age'] = np.random.randint(10, 30 + 1)

for i in range(len(data)):
    if pd.isna(data.at[i, 'Fare']):
        data.at[i, 'Fare'] = np.random.randint(7, 120 + 1)

data.info()
print(data)
```

```python
#Descartamos las columnas que tienen muchos datos perdidos o que no son muy necesarios para el analisis
data.drop(['names'], axis=1, inplace=True)
```

```python
# Dividir en conjunto de entrenamiento y prueba
train, test = train_test_split(data, test_size=0.2, random_state=42, stratify=None)

# Verificar las formas de los conjuntos resultantes
print("Forma de train:", train.shape)
print("Forma de test:", test.shape)

pd.DataFrame(train).to_csv('train.csv', index=False)
pd.DataFrame(test).to_csv('test.csv', index=False)
```
